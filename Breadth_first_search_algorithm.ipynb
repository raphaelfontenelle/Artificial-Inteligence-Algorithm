# **Breadth-first search algorithm (BFS)**

Breadth-first search is a simple strategy in which the root node is expanded first, then all the
successors of the root node are expanded next, then their successors, and so on. In general,
all the nodes are expanded at a given depth in the search tree before any nodes at the next
level are expanded.

Breadth-first search is an instance of the general graph-search algorithm in
which the shallowest unexpanded node is chosen for expansion. This is achieved very simply
by using a FIFO queue for the frontier.

# **The Algorithm**

BFS.png

```python
graph = {
  'A' : ['B','C'],
  'B' : [],
  'C' : ['D','E'],
  'D' : ['E','F'],
  'E' : ['G'],
  'F' : ['G'],
  'G' : []
}
```

```python
visited = [[]] 
queue = []   
```

```python
def bfs(visited, graph, node):
  visited.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ") 

    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
```

```python
bfs(visited, graph, 'A')
```

### **Dungeon Game problem solution**



Find the shortest path to E from S avoiding the blocks #.

Grid.png

```python
m =     [['S', '.', '.', '#', '.', '.', '.'],
        ['.','#', '.', '.', '.', '#', '.'],
        ['.','#', '.', '.', '.', '.', '.'],
        ['.', '.', '#', '#', '.', '.', '.'],
        ['.', '.', '#', 'E', '.', '#', '.']]
```

```python
# Global variabls
R, C = 5, 7 #R number of rows, C = number of colums
sr,sc = 0, 0 # Input character matrix of size R*C
rq, cq = [], [] # Empty Row Queue (rq) and Colun Queue (cq)

# Variables used to track the number of steps taken
move_count = 0
nodes_left_in_layer = 1
nodes_in_next_layer = 0

# Variable used to track whether the 'E' character
# ever gets reached during the BFS
reached_end = False

# R * C matrix of false values used to track whether
# the node at position (i, j) has been visited
visited = [[0 for i in range(7)] for j in range(5)]

# North, south, east west direction vectors.
dr = [-1, +1, 0, 0]
dc = [0, 0, +1, -1]

```

```python
def explore_neighbours(r, c):
  global nodes_in_next_layer
  for i in range(4):
    rr = r + dr[i]
    cc = c + dc[i]

    #Skip out of bounds locations
    if rr < 0 or cc < 0: continue
    if rr >= R or cc >= C: continue

    # Skip visited location or blocked cells
    if visited[rr][cc]:continue
    if m[rr][cc] == '#':continue

    rq.append(rr)
    cq.append(cc)
    visited[rr][cc] = True
    nodes_in_next_layer +=1
```

```python
def solve():
  global nodes_left_in_layer
  global nodes_in_next_layer
  global reached_end
  global move_count
  rq.append(sr)
  cq.append(sc)
  visited[sr][sc] = True
  while len(rq) > 0:
    r = rq.pop(0)
    c = cq.pop(0)
    print("({}, {}) = {}".format(r, c, m[r][c]))
    if m[r][c] == 'E':
      reached_end = True
      break
    explore_neighbours(r, c)
    nodes_left_in_layer -= 1
    if nodes_left_in_layer == 0:
      nodes_left_in_layer = nodes_in_next_layer
      nodes_in_next_layer = 0
      move_count += 1
  if reached_end:
    return move_count
  return -1
```

```python
solve()
```
