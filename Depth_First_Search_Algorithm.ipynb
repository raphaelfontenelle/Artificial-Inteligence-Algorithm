# **Depth-first search**

Depth-first search always expands DEPTH-FIRST the deepest node in the current frontier of the search tree.
SEARCH
The progress of the search is illustrated in Figure below. The search proceeds immediately
to the deepest level of the search tree, where the nodes have no successors. As those nodes
are expanded, they are dropped from the frontier, so then the search “backs up” to the next
deepest node that still has unexplored successors.


The depth-first search algorithm is an instance of the graph-search algorithm in Figure below; whereas breadth-first-search uses a FIFO queue, depth-first search uses a LIFO queue. A LIFO queue means that the most recently generated node is chosen for expansion. This must be the deepest unexpanded node because it is one deeper than its parent—which, in turn, was the deepest unexpanded node when it was selected.

## **The Algorithm**

image.png

```python
graph = {
    'A' : ['B','C'],
    'B' : [],
    'C' : ['D', 'E'],
    'D' : ['E', 'F'],
    'E' : ['G'],
    'F' : ['G'],
    'G' : []
}
```

```python
visited = set() 

def dfs(visited, graph, node):
    if node not in visited:
        print (node, end = " ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
```

```python
dfs(visited, graph, 'A')
```
